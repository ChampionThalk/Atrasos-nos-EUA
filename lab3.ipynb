```r
getStats = function(input, pos){
        input %>%
          drop_na() %>%
          filter(str_starts(DESTINATION_AIRPORT, "1", negate = TRUE)) %>%
          group_by(DESTINATION_AIRPORT) %>%
          summarise(voos_bloco=n(),soma_bloco=sum(ARRIVAL_DELAY)) %>%
          ungroup()
}

blocos =read_csv_chunked("flights.csv.zip",chunk_size = 1e6,
             callback= DataFrameCallback$new(getStats),
             col_types = cols_only(DESTINATION_AIRPORT ='c',
                                         ARRIVAL_DELAY ='i'))
```

```r

atrasos <- blocos %>% group_by(DESTINATION_AIRPORT) %>% summarise(n=sum(voos_bloco),soma=sum(soma_bloco), media_correta=soma/n) %>% arrange(DESTINATION_AIRPORT)



```
